```python
import pandas as pd
```

```python
# Cargar el archivo CSV
datos = pd.read_csv("DatosA&B11-Proy3.csv")
```

```python
# Lista de columnas a eliminar
columnas_eliminar = ["estu_consecutivo", "cole_cod_dane_establecimiento", "cole_cod_dane_sede", 
                     "cole_cod_depto_ubicacion", "cole_cod_mcpio_ubicacion", "cole_codigo_icfes", 
                     "cole_nombre_establecimiento", "cole_nombre_sede","cole_sede_principal", "estu_cod_depto_presentacion",
                     "estu_cod_mcpio_presentacion", "estu_cod_reside_depto", "estu_cod_reside_mcpio", "cole_depto_ubicacion", "estu_depto_reside",
                     "estu_estudiante","estu_estadoinvestigacion", "estu_mcpio_reside", "estu_fechanacimiento", "estu_genero", "fami_cuartoshogar", 
                     "fami_educacionmadre", "estu_pais_reside", "fami_educacionpadre", "fami_personashogar", "cole_mcpio_ubicacion", "estu_mcpio_presentacion"]

# Eliminar las columnas
datos_limpios = datos.drop(columnas_eliminar, axis=1)
```

```python
# Seleccionar solo las columnas de interés
columnas_interes = ["desemp_ingles", "punt_ingles", "punt_matematicas", 
                    "punt_sociales_ciudadanas", "punt_c_naturales", 
                    "punt_lectura_critica", "punt_global"]

# Eliminar filas con valores nulos en las columnas de interés
datos_limpios = datos_limpios.dropna(subset=columnas_interes, how='any')

# Verificar la cantidad de registros después de eliminar las filas con valores nulos
print(len(datos_limpios))
```

```python
# Verificar si hay valores nulos en todo el dataframe limpio
print(datos_limpios.isnull().sum())
```

```python
# Eliminar filas con valores nulos en todo el dataframe limpio
datos_limpios = datos_limpios.dropna()

# Verificar la cantidad de registros después de eliminar filas con valores nulos
print(len(datos_limpios))
```

```python
# Tipos de documento que queremos mantener
tipos_documento_validos = ['TI', 'CC', 'CE', 'PEP']

# Filtrar el dataframe para mantener solo los registros con los tipos de documento válidos y cole_calendario diferente de 'OTRO'
datos_filtrados = datos_limpios[(datos_limpios['estu_tipodocumento'].isin(tipos_documento_validos)) & 
                                (datos_limpios['cole_calendario'] != 'OTRO') &
                                (datos_limpios['cole_caracter'] != 'NO APLICA') &
                                (datos_limpios['fami_estratovivienda'] != "Sin Estrato")]

# Verificar la cantidad de registros después del filtrado
print(len(datos_filtrados))
```
